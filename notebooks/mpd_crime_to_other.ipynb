```python
import os
os.chdir("..")
import pandas as pd

outcomes = pd.read_csv("data/outcomes_concat.csv")
stns = pd.read_csv("data/stop_and_search_concat.csv")
street = pd.read_csv("data/street_concat.csv")
```

```python
street = street.drop(columns=['Unnamed: 0'])
street.head()
```

```python
import re
street.loc[~street["LSOA name"].isna(), "Borough"] = street["LSOA name"].dropna().apply(lambda x: re.split("[0-9]", x)[0].strip())
london_bouroughs = ['Barking and Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich', 'Hackney', 'Hammersmith and Fulham', 'Haringey', 'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington', 'Kensington and Chelsea', 'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 'Newham', 'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets', 'Waltham Forest', 'Wandsworth', 'City of Westminster', 'Richmond Upon Thames']
street = street[street["Borough"].isin(london_bouroughs)]
```

```python
asbs = street[street["Crime ID"].isna()].copy()
non_asb = street[~street["Crime ID"].isna()].copy()
```

```python
def qmap(x):
    return (max(x, 1) -1)//3 + 1

def split_ymq(df):
    df["Year"] = df["Month"].apply(lambda x: x.split("-")[0])
    df["Month"] = df["Month"].apply(lambda x: x.split("-")[1])
    df["Quarter"] = df["Month"].apply(lambda x: qmap(int(x)))

split_ymq(non_asb)
split_ymq(asbs)

non_asb.head()
```

```python
def count_both(df_a, df_na, x):
    as_ = df_a.groupby(["Year", x, "Borough", "Crime type"])["Location"].count()
    ns = df_na.groupby(["Year", x, "Borough", "Crime type"])["Crime ID"].unique().map(len)

    return pivot_onto_y_x_b(as_, ns, x)

def pivot_onto_y_x_b(df_a, df_na, x):
    ns = df_na.reset_index()
    a = df_a.reset_index()
    ns = ns.pivot_table(index=["Year", x, "Borough"], columns="Crime type", values="Crime ID", fill_value=0)
    a = a.pivot_table(index=["Year", x, "Borough"], columns="Crime type", values="Location", fill_value=0)
    ns.columns = [f'{col}' for col in ns.columns]
    a.columns = [f'{col}' for col in a.columns]
    ns, a = ns.align(a, axis=1, fill_value=0)
    ns = ns.add(a, fill_value=0)
    ns = ns.reset_index()
    return ns

month_crime = count_both(asbs, non_asb, "Month")
quarter_crime = count_both(asbs, non_asb, "Quarter")
```

```python
month_crime.to_csv("data/month_crime.csv")
quarter_crime.to_csv("data/quarter_crime.csv")
```
